### 2.1. Case Folding (Transform Comment Text into Lowercase Format)

Text preprocessing pertama yang akan saya lakukan di dalam penelitian kali ini adalah melakukan case folding terhadap seluruh teks komentar yang berhasil diambil dari video debat capres - cawapres dari babak pertama hingga babak kelima. Proses case folding sangat sering untuk dilakukan di dalam penelitian analisis sentimen karena tujuannya adalah membuat suatu format kalimat di dalam kolom teks menjadi seragam. Karena jika kita mengambil sebuah data dari resource eksternal (dalam kasus ini data-data yang diambil merupakan data komentar dari 5 konten video YouTube dengan 3 channel berbeda), maka data-data teks komentar yang diambil tersebut memiliki format penulisan yang acak, tidak seragam. Maka dari itu agar format penulisannya menjadi seragam antara 1 komentar dengan komentar lainnya, saya akan terlebih dahulu melakukan proses case folding.

```python
"""
Mengubah setiap huruf dan kata yang terdapat di dalam kolom "text_comment"
menjadi seragam yaitu setiap kalimat akan menjadi huruf kecil dan perubahan
seluruh kalimat menjad format penulisan yang seragam diletakkan di dalam kolom baru yaitu "text_comment_lower"
"""
comments_debate_df["text_comment_lower"] = comments_debate_df["text_comment"].str.lower()
comments_debate_df.head(8)
```

### 2.2. Removal Punctuations

Text preprocessing selanjutnya yang akan saya lakukan di dalam penelitian kali ini adalah membersihkan isi teks komentar dari setiap tanda baca/karakter spesial/emoticon yang terdapat di dalam kolom "text_comment" seperti, !()*+,-./:;<=>?@[]^_`{|}~]"#$%&'. Alasannya adalah karena karakter spesial/simbol-simbol khusus tidak dibutuhkan selama proses analisa sentimen dan pelabelan berlangsung.

```python
"""
Menghilangkan/menghapus simbol, karakter spesial atau emoticon yang terdapat di dalam
"text_comment_lower" (karena saat ini format penulisan dari setiap teks komentar
telah diseragamkan dan disimpan di dalam kolom "text_comment_lower") karena simbol/karakter spesial tidak dibutuhkan
selama proses penelitian analisa sentimen serta agar teks komentar yang dihasilkan menjadi lebih rapi dan bersih
"""

def cleansing_text(comment_text):
    comment_text = re.sub(r"[^a-zA-z0-9]", " ", str(comment_text))
    comment_text = re.sub(r"\b\w{1,2}\b", "", comment_text)
    comment_text = re.sub(r"\s\s+", " ", comment_text)
    comment_text = re.sub(r"https?:\/\/\S+", "", comment_text)  
    comment_text = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", comment_text)
    return comment_text

comments_debate_df["text_comment_lower"] = comments_debate_df["text_comment_lower"].apply(cleansing_text)

comments_debate_df.head(7) # Menampilkan 7 data teratas yang dimiliki oleh variabel DataFrame "comments_debate_df"
```

Setelah melakukan proses pembersihan komentar pada kolom "text_comment_lower" terhadap simbol-simbol atau karakter spesial atau bahkan segala jenis emoticon yang terdapat di dalam teks komentar, langkah selanjutnya adalah saya akan melakukan proses tokenization. Tokenization merupakan tahap pemotongan suatu kalimat utuh menjadi beberapa part/satuan kata per kata atau bisa dikatakan memotong susunan kalimat/teks utuh menjadi struktur kata per kata yang membentuk susunan kalimat utama tersebut. Tahapan tokenization sangat sering dijumpai dalam melakukan pengolahan data dokumen atau nilai data yang bertipe data String seperti analisa sentimen.

### 2.3. Tokenization (split complete text into multiple text sections)

```python
"""
Mentransformasikan/mengubah struktur susunan kalimat/teks utuh yang terdapat
di dalam kolom "text_comment_lower" menjadi potongan kata per kata di dalam list dan dipisahkan oleh
tanda koma yang di mana kata per kata tersebut jika digabungkan akan membentuk susunan kalimat utuh serta
hasil tokenization disimpan di dalam kolom baru yakni "text_comment_tokenization"
dengan menggunakan word_tokenize dari library nltk (Natural Language Tool-Kit)
"""

def text_comment_tokenization(comment_text):
    return word_tokenize(comment_text)

comments_debate_df["text_comment_tokenization"] = comments_debate_df["text_comment_lower"].apply(text_comment_tokenization)
comments_debate_df.tail(3) # Menampilkan 3 data terbawah yang dimiliki oleh variabel DataFrame "comment_debate_df"
```

Setelah melakukan proses tokenization, selanjutnya saya akan melakukan proses stopwords. Proses stopwords bertujuan untuk menghilangkan kata yang tidak memiliki arti khususnya kata Berbahasa Indonesia serta hanya menyisakan kata-kata penting saja. Tahap text processing selanjutnya yang akan digunakan di dalam penelitian kali ini adalah proses stemming. Stemming merupakan sebuah pemrosesan di dalam NLP (Natural Language Processing) yang merupakan serangkaian proses/tahapan di dalam text preprocessing yang bertujuan untuk menghilangkan kata yang memiliki imbuhan, baik kata dengan imbuhan diawal maupun kata dengan imbuhan diakhir, menjadi kata dasar tanpa memiliki imbuhan apapun. Dengan dilakukan proses stemming di suatu penelitian yang menggunakan metode NLP, maka data teks yang dimiliki dapat terstandarisasi secara baik karena kata yang dihasilkan menjadi lebih sederhana dan baku sehingga dapat membantu dalam meningkatkan performa dalam mengklasifikasikan sentimen dalam ke dalam 3 label. Sebagai contoh di dalam suatu susunan kalimat terdapat sebuah teks seperti "aku berjalan di taman kota dengan sahabatku". Maka setelah dilakukan proses stemming, susunan kalimat tersebut akan bertranformasi menjadi seperti "aku jalan di taman kota dengan sahabat".

Jadi menurut saya proses stemming di dalam penelitian yang memanfaatkan metode Natural Language Processing wajib dilakukan untuk meringkas suatu kalimat dengan mengubah kata di dalam kalimat tersebut yang memiliki imbuhan menjadi kata dasarnya.

### 2.5. Stemming (transform the format of words that have affixes into their base words)

```python
"""
Mendeklarasikan serangkaian proses stemming dengan cara menampilkan setiap perubahan kata per kata,
baik kata yang memiliki imbuhan atau kata yang tidak memiliki imbuhan (prefix maupun suffix)
dari seluruh teks komentar yang terdapat di dalam kolom "text_comment_stopword"
"""

# Membuat variabel bernama "stemmer_factory" yang menampung fungsi Stemming dari library Sastrawi yang bernama StemmerFactory
stemmer_factory = StemmerFactory()
stemmer = stemmer_factory.create_stemmer()


"""
Membuat fungsi stemming dengan 1 nilai parameter yaitu comment_text dengan tujuan setiap
kata yang memiliki imbuhan, baik itu prefix/suffix, akan langsung diubah menjadi format kata dasarnya
"""
def stemming_comment_text(comment_text):
    return stemmer.stem(comment_text)


"""
Membuat sebuah dictionary untuk menampung proses stemming. Jadi setiap hasil perubahan kata
berimbuhan (prefix maupun suffix) menjadi kata dasar, akan disimpan di dalam variabel
dictionary yang bernama stem_into_base_word
"""
stem_into_base_word = {}


# Menghitung jumlah kata dari seluruh kalimat teks komentar yang tersimpan di dalam kolom "text_comment_stopword"
for document in comments_debate_df["text_comment_stopword"]:
    for comment_text in document:
        if comment_text not in stem_into_base_word:
            stem_into_base_word[comment_text] = " "


"""
Menghitung jumlah kata yang dimiliki dan tersimpan di dalam variabel
dictionary "stem_into_tase_word" serta menampilkan jumlah kata yang dimiliki
dan tersimpan di dalam dictionary
"""
print("Jumlah Kata berdasarkan Seluruh Kalimat Teks Komentar yang telah Dilakukan Proses Stopwords adalah:",
      len(stem_into_base_word))
print("-------------------------------")


"""
Mengaplikasikan serta menampilkan progress 1 per 1 setiap perubahan kata,
baik kata yang memiliki imbuhan ataupun tidak
"""
for comment_text in stem_into_base_word:
    stem_into_base_word[comment_text] = stemming_comment_text(comment_text)
    print(comment_text, ":", stem_into_base_word[comment_text])
    
# Menampilkan seluruh kata, baik kata berimbuhan prefiks/suffix, yang telah diubah menjadi bentuk kata dasar
print(stem_into_base_word)


"""
Sebuah fungsi untuk menerapkan perubahan dari setiap kata yang memiliki imbuhan (prefix atau suffix)
menjadi kata dasarnya yang nantinya fungsi ini akan diaplikasikan ke dalam kolom "text_comment_stemming"
dari variabel DataFrame "comments_debate_df"
"""
def stemming_words_in_comment(comment):
    return [stem_into_base_word[comment_text] for comment_text in comment]


"""
Mengaplikasikan setiap perubahan kata yang memiliki imbuhan diawal maupun diakhir berdasarkan
seluruh teks komentar dari kolom "text_comment_stopword" dan perubahan disimpan di dalam kolom
baru yakni "text_comment_stemming"
"""
comments_debate_df["text_comment_stemming"] = comments_debate_df[
    "text_comment_stopword"].swifter.apply(stemming_words_in_comment)
```

Seperti kata "kayaknya" yang merupakan kata imbuhan dari kata dasar yakni "kayak" atau kata "contekan" yang merupakan kata berimbuhan dari kata "contek" sebagai kata dasarnya.

Di samping itu, setelah melewati beberapa tahapan dalam text preprocessing untuk teks komentar yang dimulai dari proses case folding yakni mengubah format penulisan dari yang sebelumnya acak dan tidak terstruktur menjadi kalimat dengan format penulisan yang seragam dengan seluruh katanya merupakan huruf kecil, membersihkan teks komentar dari simbol-simbol/karakter spesial/emoticon yang terdapat di dalam teks komentar, melakukan proses tokenizing, stopwords removal dan stemming, tahap selanjutnya di dalam text preprocessing ini adalah saya akan menggabungan hasil dari seluruh proses text processing yang saat ini disimpan di dalam kolom "text_comment_stemming" ke dalam sebuah kolom baru yang bernama "clean_text_comment". Tujuan dilakukannya pengaplikasian ini tentu saja agar membuat kata per kata yang sebelumnya terpisah dengan tanda koma dan berada di dalam suatu list dikarenakan teks komentar telah melewati tahapan tokenization (pemisahan keutuhan susunan kalimat menjadi potongan kata per kata yang dipisahkan oleh tanda koma) menjadi susunan kalimat yang utuh kembali tanpa adanya pemisahan/potongan kata per kata seperti yang dilakukan jika mengaplikasikan proses tokenization di dalam suatu kalimat.

```python
"""
Memisahkan 48.239 data komentar yang terdapat di dalam DataFrame "comments_debate_df"
menjadi terpisah dan masing-masing DataFrame akan memiliki jumlah data sebanyak 10.000 komentar
"""
def split_dataframe_comments_debate(comments_debate_df_split, chunk_size = 10000):
    chunks = list()
    number_of_chunks = len(comments_debate_df_split) // chunk_size + 1
    for i in range(number_of_chunks):
        chunks.append(comments_debate_df[i * chunk_size:(i + 1) * chunk_size])
    return chunks

comments_debate_df_split = split_dataframe_comments_debate(comments_debate_df)


"""
Menampilkan jumlah data dan kolom yang dimiliki oleh masing-masing DataFrame
hasil pemisahan dari induk DataFrame "comments_debate_df" yang masing-masing
DataFrame hasil pemisahan tersebut memiliki jumlah data sebanyak 10.000 komentar
(sesuai dengan nilai yang dibuat pada parameter chunk_size pada fungsi split_dataframe_comments_debate)
"""
print("Jumlah Baris dan Kolom yang Dimiliki oleh Turunan DataFrame Pertama Hasil Pemisahan DataFrame 'comments_debate_df'",
      comments_debate_df_split[0].shape, "\n")

print("Jumlah Baris dan Kolom yang Dimiliki oleh Turunan DataFrame Kedua Hasil Pemisahan DataFrame 'comments_debate_df'",
      comments_debate_df_split[1].shape, "\n")

print("Jumlah Baris dan Kolom yang Dimiliki oleh Turunan DataFrame Ketiga Hasil Pemisahan DataFrame 'comments_debate_df'",
      comments_debate_df_split[2].shape, "\n")

print("Jumlah Baris dan Kolom yang Dimiliki oleh Turunan DataFrame Keempat Hasil Pemisahan DataFrame 'comments_debate_df'",
      comments_debate_df_split[3].shape, "\n")

print("Jumlah Baris dan Kolom yang Dimiliki oleh Turunan DataFrame Kelima Hasil Pemisahan DataFrame 'comments_debate_df'",
      comments_debate_df_split[4].shape)
```

Sebagai pengingat saja bahwasanya melakukan proses pemisahan DataFrame menjadi lima seperti penelitian ini adalah agar resource dan waktu pemrosesan dalam memberikan skor subjectivity dan polarity terhadap masing-masing teks komentar tidak lama sehingga sangat mungkin untuk dilakukan. Hal tersebut dikarenakan variabel DataFrame utama memiliki jumlah data 48.239 sehingga jika dilakukan proses penilaian subjectivity dan polarity ke dalam 48.239 data secara langsung akan membutuhkan waktu penilaian yang sangat lama.

Maka dari itu proses pemisahan 48.239 data komentar ke dalam 5 variabel DataFrame harus dilakukan agar dapat melakukan proses penilaian subjectivity dan polarity.

```python
"""
Membuat penilaian skor subjektivitas berdasarkan isi dari seluruh teks komentar yang telah berhasil di-scraping
dari YouTube channel Najwa Shihab, KOMPASTV dan KPU RI terkait debat capres - cawapres tahun 2024 yang saat ini
tersimpan di dalam 5 variabel DataFrame yaitu "comments_debate_df_split[0]", "comments_debate_df_split[1]", dst
"""
def getSubjectivity_comments(video_comment):
    comment_text_analysis = TextBlob(video_comment)
    try:
        return comment_text_analysis.translate(from_lang="id", to="en").sentiment.subjectivity
    except:
        return comment_text_analysis.sentiment.subjectivity


"""
Membuat atribut baru yang bernama "subjectivity" yang bertujuan untuk menyimpan
hasil fungsi yang dilakukan oleh getSubjectivity_comments yang didalamnya telah dilakukan
proses translate/penerjemahan dari bahasa indonesia menjadi bahasa inggris
karena library TextBlob masih belum dapat menerjemahkan kalimat Berbahasa Indonesia
serta memberikan nilai/skor subjectivity untuk seluruh teks komentar yang tersimpan di dalam DataFrame
"comments_debate_df_split[0]", "comments_debate_df_split[1]", "comments_debate_df_split[2]",
"comments_debate_df_split[3]" dan "comments_debate_df_split[4]"
"""
comments_debate_df_split[0]["subjectivity"] = comments_debate_df_split[0]["clean_text_comment"].apply(getSubjectivity_comments)
comments_debate_df_split[1]["subjectivity"] = comments_debate_df_split[1]["clean_text_comment"].apply(getSubjectivity_comments)
comments_debate_df_split[2]["subjectivity"] = comments_debate_df_split[2]["clean_text_comment"].apply(getSubjectivity_comments)
comments_debate_df_split[3]["subjectivity"] = comments_debate_df_split[3]["clean_text_comment"].apply(getSubjectivity_comments)
comments_debate_df_split[4]["subjectivity"] = comments_debate_df_split[4]["clean_text_comment"].apply(getSubjectivity_comments)
```

```python
"""
Membuat penilaian skor polarity berdasarkan isi dari seluruh teks komentar yang telah berhasil di-scraping
dari YouTube channel Najwa Shihab, KOMPASTV dan KPU RI terkait debat capres - cawapres tahun 2024 yang saat ini
tersimpan di dalam 5 variabel DataFrame hasil pemisahan yang masing-masing kelima DataFrame menyimpan 10.000 data
yaitu "comments_debate_df_split[0]", "comments_debate_df_split[1]",
"comments_debate_df_split[2]", "comments_debate_df_split[3]" dan comments_debate_df_split[4]
"""
def getPolarity_comments(video_comment):
    comment_text_analysis = TextBlob(video_comment)
    try:
        return comment_text_analysis.translate(from_lang="id", to="en").sentiment.polarity
    except:
        return comment_text_analysis.sentiment.polarity

    
"""
Membuat atribut baru yang bernama "polarity" yang bertujuan untuk menyimpan
hasil fungsi yang dilakukan oleh getPolarity_comments yang didalamnya telah dilakukan
proses translate/penerjemahan dari bahasa indonesia menjadi bahasa inggris
karena library TextBlob masih belum dapat menerjemahkan kalimat Berbahasa Indonesia
serta memberikan nilai/skor subjectivity untuk seluruh teks komentar yang tersimpan di dalam DataFrame
"comments_debate_df_split[0]", "comments_debate_df_split[1]", "comments_debate_df_split[2]",
"comments_debate_df_split[3]" dan "comments_debate_df_split[4]"
"""
comments_debate_df_split[0]["polarity"] = comments_debate_df_split[0]["clean_text_comment"].apply(getPolarity_comments)
comments_debate_df_split[1]["polarity"] = comments_debate_df_split[1]["clean_text_comment"].apply(getPolarity_comments)
comments_debate_df_split[2]["polarity"] = comments_debate_df_split[2]["clean_text_comment"].apply(getPolarity_comments)
comments_debate_df_split[3]["polarity"] = comments_debate_df_split[3]["clean_text_comment"].apply(getPolarity_comments)
comments_debate_df_split[4]["polarity"] = comments_debate_df_split[4]["clean_text_comment"].apply(getPolarity_comments)

#comments_debate_df_split[0].head(2) # Menampilkan 14 data teratas setelah kolom baru yang bernama "subjectivity" dibuat
```

```python
"""
Melakukan proses labelling otomatis dengan library textblob yang dimana
label pada eksperimen ini terdiri dari 3: Negatif, Netral, dan Positif
"""

def sentiment_label(score):
    if score < 0:
        result_sentiment_label = "negatif"
    elif score == 0:
        result_sentiment_label = "netral"
    else:
        result_sentiment_label = "positif"
        
    return result_sentiment_label
    

comments_debate_df_split[0]["sentiment_label"] = comments_debate_df_split[0]["polarity"].apply(sentiment_label)
comments_debate_df_split[1]["sentiment_label"] = comments_debate_df_split[1]["polarity"].apply(sentiment_label)
comments_debate_df_split[2]["sentiment_label"] = comments_debate_df_split[2]["polarity"].apply(sentiment_label)
comments_debate_df_split[3]["sentiment_label"] = comments_debate_df_split[3]["polarity"].apply(sentiment_label)
comments_debate_df_split[4]["sentiment_label"] = comments_debate_df_split[4]["polarity"].apply(sentiment_label)
```
