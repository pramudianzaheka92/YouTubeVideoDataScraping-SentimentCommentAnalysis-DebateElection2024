Kesepuluh indeks baris data tersebut memiliki nilai NaN atau Not a Number, sebuah nilai yang merepresentasikan bahwa data tersebut tidak memiliki nilai alias missing value.

Seperti yang sudah diinformasikan sebelumnya bahwasanya dalam penelitian kali ini, saya akan menggantikan nilai-nilai NaN yang dimiliki oleh kolom "author" dan "channel_url" menjadi nilai kategorik "Unknown" supaya data-data yang terkandung di kolom lainnya tidak terbuang sehingga dapat mempertahankan keaslian informasi dan nantinya akan sangat berguna untuk Algoritma Machine Learning dalam mempelajari pola-pola serta juga akan sangat berguna bagi tahapan text processing seperti tokenization, stopwords, dll., karena akan mempelajari banyak pola teks Bahasa Indonesia.

```python
"""
Mentransformasikan nilai-nilai data yang hilang (missing value)
pada kolom "author" dan "channel_url" yang dimiliki variabel DataFrame "data_kpu_ri"
dengan sebuah nilai string/kategorik yang ditentukan (spesific string value) yaitu Unknown
untuk setiap data yang sebelumnya memiliki nilai NaN pada baris data tersebut
"""

def change_missing_value_data(data_kpu_ri):
    """
    Mengubah missing value di dalam kolom "author" menjadi spesifik string yakni "Unknown"
    sehingga dapat diasumsikan jika nama user/akun YouTube-nya tidak diketahui
    """
    data_kpu_ri["author"] = data_kpu_ri["author"].fillna("Unknown")
    
    """
    Mengubah missing value di dalam kolom "channel_url" juga menjadi spesifik string yakni "Unknown"
    sehingga dapat diasumsikan jika url/link yang mengarahkan ke profile user/akun tidak diketahui
    """
    data_kpu_ri["channel_url"] = data_kpu_ri["channel_url"].fillna("Unknown")
    
    
    return change_missing_value_data

change_missing_value_data(data_kpu_ri)

# Menampilkan kembali jumlah nilai data yang hilang dari masing-masing kolom pada variabel DataFrame "data_kpu_ri"
print("Jumlah Nilai Data yang Hilang (Missing Value) dari Masing-Masing Kolom setelah Missing Value Dihilangkan adalah:")
data_kpu_ri.isnull().sum()
```

Berdasarkan output diatas diketahui bahwasanya masing-masing kolom khususnya kolom "author" dan kolom "channel_url" sudah tidak memiliki nilai NaN yang menjadi representasi bahwa pada data tersebut memiliki nilai data yang hilang (missing value).

Sebagai penjelasan tambahan saat saya melakukan proses penggabungan ketiga variabel DataFrame menjadi sebuah DataFrame yang tersentralisasi dengan nama DataFrame yaitu "comments_debate_df", saya menggunakan sebuah parameter di dalam fungsi pandas pd.concat() yaitu parameter yang bernama ignore_index yang bernilai True (sebuah nilai dari tipe data Boolean). Parameter tersebut saya gunakan (ignore_index = True) bertujuan agar indeks pada DataFrame "comments_debate_df" tidak terduplikasi dan tidak mengulangi nilai indeks yang sama dari indeks awal.
