```python
def video_youtube_comments(video_id):
    youtube_video_comments = [] # Melakukan proses inisialisasi list kosong untuk menyimpan data-data komentar dari video youtube
    
    youtube = build("youtube", "v3", developerKey = api_key)
    
    video_response = youtube.commentThreads().list(part = "snippet,replies",
                                                   videoId = video_id).execute()

    while video_response:
        for item in video_response["items"]:
            
            channel_video = item["snippet"]["topLevelComment"]["snippet"]["channelId"]
            author_channel_link = item["snippet"]["topLevelComment"]["snippet"]["authorChannelUrl"]
            user = item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
            date_comment_publish = item["snippet"]["topLevelComment"]["snippet"]["publishedAt"]
            date_updated_comment = item["snippet"]["topLevelComment"]["snippet"]["updatedAt"]
            like_count = item["snippet"]["topLevelComment"]["snippet"]["likeCount"]
            text_comment_video = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            
            youtube_video_comments.append([channel_video, author_channel_link, user,
                                           date_comment_publish, date_updated_comment, like_count, text_comment_video])
            
            replycount = item["snippet"]["totalReplyCount"]
            
            if replycount > 0:
                for reply in item["replies"]["comments"]:
                    date_comment_publish = reply['snippet']['publishedAt']
                    user = reply['snippet']['authorDisplayName']
                    text_reply = reply['snippet']['textDisplay']
                    likeCount = reply['snippet']['likeCount']
                    
                    youtube_video_comments.append([date_comment_publish, user, text_reply, likeCount])
            
            
        if "nextPageToken" in video_response:
            video_response = youtube_first_channel_object.commentThreads().list(
                part = "snippet, replies",
                pageToken = video_response["nextPageToken"],
                videoId = video_id
            ).execute()
        else:
            break
        return youtube_video_comments
```
