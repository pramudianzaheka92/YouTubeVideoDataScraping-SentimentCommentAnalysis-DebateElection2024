## 2. Combining Five DataFrame Variables Contain Comments Information from the First - Fifth Debate Videos on KOMPASTV YouTube Channel

```python
"""
Menggabungkan 5 variabel DataFrame menjadi 1 file dataset dalam format
spreadsheet Microsoft Excel dimana masing-masing variabel DataFrame memiliki jumlah
data yang disimpan berbeda-beda karena saat jumlah data yang didefinisikan saat
melakukan proses scraping/pengambilan informasi-informasi komentar utama dengan metode pengulangan (looping)
dibuat berbeda antara jumlah data komentar dari video debat pertama, debat kedua, dst
"""

df_comment_video_debate_kompastv = pd.concat([first_debate_video_df, second_debate_video_df, third_debate_video_df,
                                              fourth_debate_video_df, fifth_debate_video_df], ignore_index = True)

# Menampilkan 35 data teratas yang dimiliki oleh variabel DataFrame "df_comment_video_debate_kompastv"
df_comment_video_debate_kompastv.head(35)
```

Berdasarkan code diatas, saya telah berhasil menggabungkan 5 variabel DataFrame yang masing-masing DataFrame memiliki data/informasi komentarnya masing-masing. Sebagai contoh pada variabel DataFrame "first_debate_video_df" berisikan 10.500 data/informasi komentar yang berhasil diambil dari video debat babak pertama capres-cawapres Republik Indonesia tahun 2024 yang ditayangkan oleh YouTube channel KOMPASTV. Sedangkan variabel DataFrame "second_debate_video_df" juga memiliki data/informasi komentar utama namun yang membedakannya dengan DataFrame "first_debate_video_df" adalah jumlah data yang diambil dan sumber videonya karena jika variabel DataFrame "second_debate_video_df" memiliki 500 dan sumber informasi/data komentar yang diambil berasal dari video debat babak kedua capres-cawapres Republik Indonesia tahun 2024 yang di-upload oleh YouTube channel KOMPASTV.

Hal tersebut juga sama dengan yang dimiliki oleh variabel DataFrame lainnya seperti "third_debate_video_debate_df", "fourth_debate_video_df" dan "fifth_debate_video_df" dengan catatan jumlah data yang diambil berbeda-beda dan sumber video atas informasi komentar yang diambil juga berbeda namun video-video tersebut tetap di-upload/ditayangkan oleh YouTube channel KOMPASTV.

Untuk mengecek dan memastikan apakah penggabungan 5 variabel DataFrame berhasil dilakukan, saya akan menampilkan 11 data terbawah yang dimiliki oleh DataFrame "df_comment_video_debate_kompastv" karena jika proses penggabungan 5 variabel DataFrame menjadi 1 DataFrame berhasil dilakukan maka 11 data terbawah seharusnya berasal dari data-data komentar yang diambil (di-scraping) dari video debat babak kelima capres-cawapres Republik Indonesia tahun 2024 yang ditayangkan oleh YouTube channel KOMPASTV dan disimpan di dalam variabel DataFrame bernama "fifth_debate_video_df".

Sebagai penjelasan tambahan saat saya melakukan proses penggabungan 5 variabel DataFrame yang berbeda-beda menjadi 1 DataFrame agar data-data komentar dari video debat babak pertama hingga babak kelima yang ditayangkan oleh channel YouTube KOMPASTV tidak terpisah-pisah dan tersentralisasi (terpusat) ke dalam 1 file dataset, saya menggunakan sebuah parameter yaitu ignore_index yang bernilai Boolean yaitu True. Parameter tersebut saya gunakan (ignore_index = True) bertujuan agar indeks pada DataFrame (dalam hal ini indeks yang terdapat di dalam variabel DataFrame "df_comment_video_debate_kompastv") tidak terduplikasi dan tidak mengulangi nilai indeks yang sama dari indeks awal.

Sebagai contoh variabel DataFrame "first_video_debate_df" memiliki sebuah nilai awal indeks yaitu 0 (indeks data 0 mengartikan jika data tersebut merupakan baris pertama yang ada di dalam data tersebut) dan indeks terakhir pada variabel DataFrame "first_video_debate_df" adalah 10499 yang mengartikan bahwa baris data tersebut merupakan baris data ke-10.500. Lalu karena DataFrame "first_video_debate_df" digabungkan dengan DataFrame "second_video_debate_df" maka baris data selanjutnya sudah pasti merupakan data yang dimiliki oleh indeks data ke-0/baris pertama dari DataFrame "second_video_debate_df".

Jika tidak menggunakan parameter ignore_index yang bernilai True, maka variabel DataFrame "df_comment_video_debate_kompastv" yang telah menjadi DataFrame tersentralisasi tidak melanjutkan nilai indeks yang selanjutnya yaitu indeks data ke-10500 yang merupakan data pertama/indeks data ke-0 dari variabel DataFrame "second_video_debate_df". Jadi secara tidak langsung akan ada 2 baris data yang memiliki indeks ke-0 jika DataFrame "df_comment_video_debate_kompastv" berisikan data-data komentar dari DataFrame "first_video_debate_df" dan "second_video_debate_df". Karena DataFrame yang tersentralisasi ini berisikan data-data komentar hasil penggabungan 5 variabel DataFrame yang berbeda, maka jika tidak menggunakan parameter ignore_index bernilai True DataFrame "df_comment_video_debate_kompastv" memiliki 5 baris data yang memiliki indeks data ke-0

Jadi alasan tersebutlah saya menggunakan parameter ignore_index = True untuk menghindari terjadinya duplikasi indeks data pada DataFrame yang tersentralisasi/terpusat yaitu variabel DataFrame "df_comment_video_debate_kompastv".
